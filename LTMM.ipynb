## Sử dụng vòng lặp

```python
def computeGCD(x, y):
  
    if x > y:
        small = y
    else:
        small = x
    for i in range(1, small+1):
        if((x % i == 0) and (y % i == 0)):
            gcd = i          
    return gcd
a = 60
b= 48
  
print ("Ước chung lớn nhất của a và b : ",end="")
print (computeGCD(60,48))
```

## Sử dụng Euclide mở rộng


```python
def gcdExtended(a, b): 
    # Base Case 
    if a == 0 :  
        return b,0,1
             
    gcd,x1,y1 = gcdExtended(b%a, a) 
     
    # Update x and y using results of recursive 
    # call 
    x = y1 - (b//a) * x1 
    y = x1 
     
    return gcd,x,y

# Driver code
a, b = 35,15
g, x, y = gcdExtended(a, b) 
print("gcd(", a , "," , b, ") = ", g) 
```

## encryption

```python
def caesar_encryption(plaintext,key):
  encryption_str = ''
  for i in plaintext:
    if i.isupper():
      temp = 65 + ((ord(i) - 65 + key) % 26) 
      encryption_str = encryption_str + chr(temp)                              
    elif i.islower():
      temp = 97 + ((ord(i) - 97 + key) % 26)
      encryption_str = encryption_str + chr(temp)
    else:
      encryption_str = encryption_str + i  
 
  print("The ciphertext is:",encryption_str)
 
plaintext = input("Enter the plaintext:")
key = int(input("Enter the key:"))
caesar_encryption(plaintext,key)
```

## decryption


```python
def caesar_decryption(ciphertext,key):
  decryption_str = ''
  for i in ciphertext:
    if i.isupper():
      if ((ord(i) - 65 - key) < 0):
        temp = 65 + ((ord(i) - 65 - key + 26) % 26) 
      else:
        temp = 65 + ((ord(i) - 65 - key) % 26) 
      decryption_str = decryption_str + chr(temp)                              
    elif i.islower():
      if ((ord(i) - 97 - key) < 0):
        temp = 97 + ((ord(i) - 97 - key + 26) % 26) 
      else:
        temp = 97 + ((ord(i) - 97 - key) % 26) 
      decryption_str = decryption_str + chr(temp)
    else:
      decryption_str = decryption_str + i  
 
  print("The plaintext is:",decryption_str)
 
ciphertext = input("Enter the ciphertext:")
key = int(input("Enter the key:"))
caesar_decryption(ciphertext,key)
```

```python
# Python program to implement Autokey Cipher
# Dictionary to lookup the index of alphabets
dict1 = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4,
         'F': 5, 'G': 6, 'H': 7, 'I': 8, 'J': 9,
         'K': 10, 'L': 11, 'M': 12, 'N': 13, 'O': 14,
         'P': 15, 'Q': 16, 'R': 17, 'S': 18, 'T': 19,
         'U': 20, 'V': 21, 'W': 22, 'X': 23, 'Y': 24, 'Z': 25}

dict2 = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E',
         5: 'F', 6: 'G', 7: 'H', 8: 'I', 9: 'J',
         10: 'K', 11: 'L', 12: 'M', 13: 'N', 14: 'O',
         15: 'P', 16: 'Q', 17: 'R', 18: 'S', 19: 'T',
         20: 'U', 21: 'V', 22: 'W', 23: 'X', 24: 'Y', 25: 'Z'}
# This function generates the key
def generate_key(message, key):
    i = 0
    while True:
        if len(key) == len(message):
            break
        if message[i] == ' ':
            i += 1
        else:
            key += message[i]
            i += 1
    return key

# This function returns the Cipertext
def cipherText(message, key_new):
    cipher_text = ''
    i = 0
    for letter in message:
        if letter == ' ':
            cipher_text += ' '
        else:
            x = (dict1[letter]+dict1[key_new[i]]) % 26
            i += 1
            cipher_text += dict2[x]
    return cipher_text
# This function decrypts the encrypted text
# and returns the original text
def originalText(cipher_text, key_new):
    or_txt = ''
    i = 0
    for letter in cipher_text:
        if letter == ' ':
            or_txt += ' '
        else:
            x = (dict1[letter]-dict1[key_new[i]]+26) % 26
            i += 1
            or_txt += dict2[x]
    return or_txt
def main():
    message = 'HOANGMINHDUC'
    key = 'DARE'
    key_new = generate_key(message, key)
    cipher_text = cipherText(message, key_new)
    original_text = originalText(cipher_text, key_new)
    print("Encrypted Text =", cipher_text)
    print("Original Text =", original_text)

# Executes the main function
if __name__ == '__main__':
    main()
```

-	Khóa K sẽ được lặp lại liên tục tùy theo chiều dài của bản mã. 

       K = [ (K1, K2, K3, K4,…), (K1, K2, K3, K4,…),…] 






```python
def generateKey(string, key): 
  key = list(key) 
  if len(string) == len(key): 
    return(key) 
  else: 
    for i in range(len(string) -len(key)): 
      key.append(key[i % len(key)]) 
  return("" . join(key)) 
  
def encryption(string, key): 
  encrypt_text = [] 
  for i in range(len(string)): 
    x = (ord(string[i]) +ord(key[i])) % 26
    x += ord('A') 
    encrypt_text.append(chr(x)) 
  return("" . join(encrypt_text)) 
def decryption(encrypt_text, key): 
  orig_text = [] 
  for i in range(len(encrypt_text)): 
    x = (ord(encrypt_text[i]) -ord(key[i]) + 26) % 26
    x += ord('A') 
    orig_text.append(chr(x)) 
  return("" . join(orig_text)) 
if __name__ == "__main__": 
  string = input("Enter the message: ")
  keyword = input("Enter the keyword: ")
  key = generateKey(string, keyword) 
  encrypt_text = encryption(string,key) 
  print("Encrypted message:", encrypt_text) 
  print("Decrypted message:", decryption(encrypt_text, key)) 
```

```python
def shift_left(k, nth_shifts):
    s = ""
    for i in range(nth_shifts):
        for j in range(1,len(k)):
            s = s + k[j]
        s = s + k[0]
        k = s
        s = ""
    return k 
```

```python
def Hien_thi_Hexa(plaintext, size):
	copy = plaintext.copy()
	copy = [ord(i)-ord('0') for i in copy] # Chuyển các kí tự '0', '1' sang dạng số
	Tong = 0
	for i in range(len(copy)): # Chuyển sang dạng thập phân
		Tong += (2**(size-i-1)) * copy[i]
	return hex(Tong)
```

Tạo dòng khóa (chuỗi các bit) và kết hợp dòng khóa với dữ liệu bản rõ sau đó sử dụng phép toán XOR để tạo dòng dữ liệu bản mã.

## Thuật toán khóa
Một khóa có độ dài thay đổi từ 1 đến 256 byte được sử dụng để khởi tạo vector trạng thái 256 byte S, với các phần tử S [0] đến S [255]. Đối với mã hóa và giải mã, một byte k được tạo ra từ S bằng cách chọn một trong 255 mục nhập theo kiểu có hệ thống, sau đó các mục trong S được hoán vị lại.

### 1. Thuật toán lập lịch khóa:
Khởi tạo : Các mục của S được đặt bằng các giá trị từ 0 đến 255 theo thứ tự tăng dần, một vectơ tạm thời T, được tạo ra.
Nếu độ dài của khóa k là 256 byte, thì k được gán cho T. Ngược lại, đối với khóa có độ dài (k-len) byte, k-len phần tử đầu tiên của T được sao chép từ K và sau đó K được lặp lại bao nhiêu lần cần thiết để điền T. 

Ý tưởng được minh họa như sau:

for
	i = 0 to 255 do S[i] = i;
T[i] = K[i mod k - len];


chúng tôi sử dụng T để tạo ra hoán vị ban đầu của S. Bắt đầu với S [0] thành S [255], và đối với mỗi thuật toán S [i] hoán đổi nó với một byte khác trong S theo một sơ đồ được chỉ định bởi T [i], nhưng S sẽ vẫn chứa các giá trị từ 0 đến 255:

j = 0;
for
	i = 0 to 255 do
	{
		j = (j + S[i] + T[i])mod 256;
		Swap(S[i], S[j]);
	}
### 2. Thuật toán tạo ngẫu nhiên giả (Stream Generation):
Khi vectơ S được khởi tạo, khóa đầu vào sẽ không được sử dụng. Trong bước này, đối với mỗi thuật toán S [i] hoán đổi nó với một byte khác trong S theo một lược đồ được chỉ định bởi cấu hình hiện tại của S. Sau khi đạt đến S [255], quá trình tiếp tục, bắt đầu lại từ S [0]

i, j = 0;
while (true)
	i = (i + 1)mod 256;
j = (j + S[i])mod 256;
Swap(S[i], S[j]);
t = (S[i] + S[j])mod 256;
k = S[t];
